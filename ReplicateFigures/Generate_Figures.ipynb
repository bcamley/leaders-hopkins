```python
def basic_dir(N,signu,h):
    return './Data/basic/N'+str(N)+'/signu'+str(signu)+'/h'+str(h)+'/'

def signu_dir(g,signu,h):
    return './Data/vary_signu/group'+str(g)+'/h'+str(h)+'/signu'+str(signu)+'/'

def N_dir(N,signu,h):
    return './Data/vary_N/signu'+str(signu)+'/h'+str(h)+'/N'+str(N)+'/'

def train_dir(N,signu,h):
    return './Data/geom/N'+str(N)+'/train/signu'+str(signu)+'/h'+str(h)+'/'

def olig_dir(N,signu,h):
    return './Data/geom/N'+str(N)+'/olig/signu'+str(signu)+'/h'+str(h)+'/'

def int_dir(I,signu,h):
    return './Data/intrinsic/signu'+str(signu)+'/h'+str(h)+'/I'+str(I)+'/'
```

```python
def C(x,h,Cmax,Cmin):
    return (Cmax-Cmin)*(1/2)*(1+np.tanh(x/h))+Cmin

def Cphi_g(phi_r,x,h,Cmax,Cmin):
    return (Cmax-Cmin)*(1/2)*(1+np.tanh((x+(1/2)*np.cos(phi_r))/h))+Cmin

def DcDphi_g(phi_r,x,h,Cmax,Cmin):
    return -(Cmax-Cmin)*(np.cosh((x+(1/2)*np.cos(phi_r))/h)**(-2))*np.sin(phi_r)/(4*h)

def FI_phi_r(phi_r,x,h,Cmax,Cmin,kd):
    con = Cphi_g(phi_r,x,h,Cmax,Cmin)
    dcon = DcDphi_g(phi_r,x,h,Cmax,Cmin)
    return (kd*dcon**2)/(con*(con+kd)**2)

def sig_deltas(x,h,Cmax,Cmin):
    FIs = []
    for i in x:
        FIs.append(Nr*integrate.quad(FI_phi_r, 0, 2*np.pi, args=(i,h,Cmax,Cmin,kd))[0]/(2*np.pi))
    FIs = np.array(FIs)
    return 1/np.sqrt(FIs)

def N_cells(Q):
    return 1+3*Q+3*Q**2

def in_row(i,Q):
    if i < 0:
        return 0
    if i >= 0 and i < Q:
        return Q+1+i
    if i == Q:
        return 2*Q+1
    if i > Q and i <= 2*Q:
        return 3*Q+1-i
    if i > 2*Q:
        return 0
    
def x_coords(Q):
    xs = []
    x = 0
    for i in range(2*Q+1):
        xs += in_row(i,Q)*[x]
        x += -1
    return np.array(xs)

def grad(x,h):
    return (1/2)*(1+np.tanh(x/h))

def signal_sig(sigma):
    conds = [sigma < 35, sigma >= 35]
    funcs = [lambda x: np.exp(-(x**2)/2)*np.real(erf((np.pi+1j*x**2)/(np.sqrt(2)*x)))/erf(np.pi/(np.sqrt(2)*x)),lambda x: 0]
    return np.piecewise(sigma,conds,funcs)

def leader_cos(leader_uncertainties,h):
    Ntrials = 100000
    if h >= 125:
        Ntrials = 500000
    trials = []
    for uncer in leader_uncertainties:
        if uncer < 60:
            a,b = -np.pi/uncer, np.pi/uncer
            trials.append(stats.truncnorm.rvs(a,b,loc=0,scale=uncer,size=Ntrials))
        if uncer >= 60:
            trials.append(np.random.uniform(-np.pi,np.pi,size=Ntrials))
    trials = np.array(trials)
    return np.mean(np.sum(np.cos(trials),axis=0)/np.sqrt(np.sum(np.cos(trials),axis=0)**2+np.sum(np.sin(trials),axis=0)**2))

def do_first_order_Vx(N,signu,h):
    delnu = signu*2*np.pi/100
    xs = np.linspace(0,-N+1,N)
    follower = signal_sig(delnu)
    leader_uncertainties = sig_deltas(xs,h,2,1e-14)
    leader_uncertainties = leader_uncertainties[leader_uncertainties.argsort()]
    Vxs = []
    Vxs.append(0)
    #print(follower)
    for j in range(1,N+1):
        Plx = np.sum(signal_sig(leader_uncertainties[:j]))
        Pfx = (N-j)*(leader_cos(leader_uncertainties[:j],h))*follower
        Vx = (1/N)*(Plx+Pfx)
        Vxs.append(Vx)
    print(signu,h)
    return np.array(Vxs),0

def do_first_order_olig_Vx(Q,signu,h):
    delnu = signu*2*np.pi/100
    N = N_cells(Q)
    xs = x_coords(Q)
    follower = signal_sig(delnu)
    leader_uncertainties = sig_deltas(xs,h,2,1e-14)
    leader_uncertainties = leader_uncertainties[leader_uncertainties.argsort()]
    Vxs = []
    Vxs.append(0)
    #print(follower)
    for j in range(1,N+1):
        Plx = np.sum(signal_sig(leader_uncertainties[:j]))
        Pfx = (N-j)*(leader_cos(leader_uncertainties[:j],h))*follower
        Vx = (1/N)*(Plx+Pfx)
        Vxs.append(Vx)
    return np.array(Vxs),0
```

## Fig 7: Representative Autocorrelation Function

```python
ts = np.load('./Data/small_data/ts.npy')
Vcs =  np.load('./Data/small_data/Vcs.npy')

def vel_correlation(Vcs):
    vcor = []
    msq = np.dot(np.mean(Vcs,axis=0),np.mean(Vcs,axis=0))
    vcor.append(np.mean(Vcs[:,0]*Vcs[:,0]+Vcs[:,1]*Vcs[:,1])-msq)
    for i in vcor_samples[1:]:
        vcor.append(np.mean((Vcs[:-i:,0]*Vcs[i::,0])+(Vcs[:-i:,1]*Vcs[i::,1]))-msq)
    return np.array(vcor)

def exp_func(x,tau):
    return np.exp(-x/tau)

N = 50
Nls = np.arange(0,N+1,1)
Nsamples = 60000
vcor_samples = np.append(np.array([0]),np.unique(np.logspace(0,np.log10(Nsamples),num=int(Nsamples/100),dtype=int)))
delta_t = 0.01

vcor = vel_correlation(Vcs[20000:])
tau = curve_fit(exp_func,vcor_samples*delta_t,vcor/vcor[0])[0][0]

fig, ax = plt.subplots(1,1)

plt.scatter(vcor_samples*delta_t,vcor/vcor[0],label='Simulation Data')
plt.plot(vcor_samples*delta_t,exp_func(vcor_samples*delta_t,tau),c='k',linestyle='--',zorder=2,label='Exponential Fit')

plt.xlabel('t')
plt.ylabel(r'$\mathbf{V}_c$ Autocorrelation')

plt.legend()

plt.xlim(-.5,25)

fig.savefig('./Figs/Vcor_exp.pdf')
```
